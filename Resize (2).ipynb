```python
# Define the path to the source folder and the destination folder
src_folder = '/content/HDDATA/vald/prowdey/'
dest_folder = '/content/drive/MyDrive/FYDATAresize/Valid/prowdey'


  
```

```python
# Define the size for the resized images
width = 640
height = 640

# Loop through the source folder and resize each image
for filename in os.listdir(src_folder):
  # Read the image
  image = cv2.imread(os.path.join(src_folder, filename))
  
  # Resize the image
  resized_image = cv2.resize(image, (width, height))
  
  # Save the resized image in the destination folder
  cv2.imwrite(os.path.join(dest_folder, filename), resized_image)
```

```python
data_folder_path = '/content/drive/MyDrive/Monkey/OUTMonkey/train/'
augmented_folder_path = '/content/drive/MyDrive/Monkey/Skin/Chickenpox/train'
```

```python
datagen = ImageDataGenerator(       #Data Augmentation Technique
    rotation_range=20,
    shear_range=0.2,
    zoom_range=0.45,
    vertical_flip=True,
    fill_mode='nearest')
```

```python
for filename in os.listdir(data_folder_path): #read folder
    img = PIL.Image.open(data_folder_path + filename) #open image dataset
    img = np.array(img)  #image convert into array
    img = img.reshape((1,) + img.shape)
    
    # Apply the augmentation techniques and save the augmented images
    i = 0
    for batch in datagen.flow(img, batch_size=20, save_to_dir=augmented_folder_path, save_prefix='aug_' + filename[:-1], save_format='jpeg'):
        i += 1
        if i > 2:
            break
```
